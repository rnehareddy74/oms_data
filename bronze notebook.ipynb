Data Reading

```python
dbutils.widgets.text("file_name","")
```

```python
m_file_name=dbutils.widgets.get("file_name")
```

```python
m_file_name
```

```python
df=spark.readStream.format("cloudFiles")\
    .option("cloudFiles.format","parquet")\
    .option("cloudFiles.schemaLocation",f"abfss://bronze@azureomsprojectusa.dfs.core.windows.net/checkpoint_{m_file_name}")\
    .load(f"abfss://source@azureomsprojectusa.dfs.core.windows.net/{m_file_name}")
```

Data writing

```python
df.writeStream.format("parquet")\
              .outputMode("append")\
              .option("checkpointLocation",f"abfss://bronze@azureomsprojectusa.dfs.core.windows.net/checkpoint_{m_file_name}")\
              .option("path",f"abfss://bronze@azureomsprojectusa.dfs.core.windows.net/{m_file_name}")\
              .trigger(once=True)\
              .start()
                  
```
